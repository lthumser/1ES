```python
%matplotlib notebook
import matplotlib.pyplot as plt

serie_1 = [1, 2, 3, 4, 10]
serie_2 = [12, 45, 4, 89, 3]
titres = ['A', 'B', 'C', 'D', 'E']

plt.subplot (1,2,1)
plt.pie(serie_1, labels = titres, labeldistance = 1.2, normalize = True, autopct='%1.0f%%', shadow = True)
plt.title("Titre sÃ©rie 1")

plt.subplot (1,2,2)
plt.pie(serie_2, labels = titres, labeldistance = 1.2, normalize = True, autopct='%1.0f%%', shadow = True)
plt.title("Titre sÃ©rie 2")

plt.legend()
```
